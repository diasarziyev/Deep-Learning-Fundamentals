# 1D-CNN for Text

```python
import numpy as np
```

```python
X = np.random.randn(1, 300, 10)
W = np.random.randn(16, 300, 3)
```

```python
def conv1d_forward_scaled(X, W, b):
    C_out = W.shape[0]
    filter_size = W.shape[2]
    length = X.shape[1]
    
    out_len = length - filter_size + 1
    conv_out = np.zeros((C_out, out_len))
    
    for f in range(C_out):
        for i in range(out_len):
            x_slice = X[:, i:i + filter_size]
            conv_out[f, i] = np.sum(x_slice * W[f]) + b[f]
    relu_out = np.maximum(0, conv_out)
    pooled_out = np.zeros(C_out)            
    max_indices = np.zeros(C_out, dtype=int) 
    
    for f in range(C_out):
        pooled_out[f] = np.max(relu_out[f])
        max_indices[f] = np.argmax(relu_out[f]) 
    cache = (X, W, b, conv_out, relu_out, max_indices)
    return pooled_out, cache
```
