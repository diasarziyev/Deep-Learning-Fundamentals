# 1D-CNN for Text

```python
import numpy as np
```

```python
X = np.random.randn(1, 300, 10)
W = np.random.randn(16, 300, 3)
```

```python
def forward_pass(X, W):
    length = X.shape[1]
    filter_size = W.shape[1]
    size = X.shape[1]
    out = []
    for i in range(0, length - filter_size + 1):
        slice = X[:, i:i + filter_size]
        out_feature = np.sum(slice * W)
        out.append(out_feature)
    out = np.array(out)
    result = np.maximum(0, out)
    result_feature = np.max(result)
    return result, result_feature
```
