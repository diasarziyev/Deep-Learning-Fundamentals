```python
import numpy as np
```

```python
def softmax(Z):
    new_Z = Z - np.max(Z, axis = 0, keepdims = True)
    exponents = np.exp(new_Z)
    return exponents / np.sum(exponents, axis = 0, keepdims = True)
```

```python
def initialize_parameters(nx, ny):
    np.random.seed(42)
    parameters = {
        'W': np.random.randn(ny, nx) * 0.01,
        'b': np.zeros((ny, 1))
    }
    return parameters
```

```python
def compute_loss(A, Y):
    m = Y.shape[1]
    A = np.clip(A, 1e-15, 1.0 - 1e-15)
    cost = - (1 / m) * np.sum(Y * np.log(A))
    return np.squeeze(cost)
```

```python
def forward_backward_pass(X, Y, parameters):
    m = Y.shape[1]
    grads = {}
    Z = np.dot(parameters['W'], X) + parameters['b']
    A = softmax(Z)
    dZ = A - Y
    grads['dW'] = (1.0 / m) * np.dot(dZ, X.T)
    grads['db'] = (1.0 / m) * np.sum(dZ, axis = 1, keepdims = True)
    return A, grads
```

```python
def update_parameters(parameters, grads, learning_rate):
    parameters['W'] -= learning_rate * grads['dW']
    parameters['b'] -= learning_rate * grads['db']
    return parameters
```

```python
def model(X, Y, ny, iter = 1000, learning_rate = 0.01):
    parameters = initialize_parameters(X.shape[0], ny)
    for i in range(iter):
        A, grads = forward_backward_pass(X, Y, parameters)
        parameters = update_parameters(parameters, grads, learning_rate)
        cost = compute_loss(A, Y)

        if i % 100 == 0:
            cost = compute_loss(A, Y)
            print(f'cost after itration {i} is {cost}')

    return parameters
```

```python
from sklearn.datasets import make_blobs
```

```python
X_raw, Y_raw = make_blobs(n_samples = 900, n_features = 2, centers = 3, random_state = 42)
X_train = X_raw.T
ny = 3 
Y_train = np.eye(ny)[Y_raw].T
```

```python
X_train.shape, Y_train.shape
```

```python
parameters = model(X_train, Y_train, ny)
```

```python
def predict(X, parameters):
    Z = np.dot(parameters['W'], X) + parameters['b']
    A = softmax(Z)
    return np.argmax(A, axis = 0)
```

```python
predictions = predict(X_train, parameters)
accuracy = np.mean(predictions == Y_raw) * 100
accuracy
```
